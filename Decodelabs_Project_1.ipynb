```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel("/content/Dataset for Data Analytics.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
numeric_cols= df.select_dtypes(include= [np.number])

for co1 in numeric_cols:
  df[co1]= df[co1].fillna(df[co1].median)
```

```python
numeric_cols= df.select_dtypes(include= ['object'])

for co1 in numeric_cols:
  df[co1]= df[co1].fillna('unknown')
```

```python
df.duplicated().sum()
```

```python
df= df.drop_duplicates()
```

```python
print(df.columns)
```

```python
df['Date']= pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)
```
